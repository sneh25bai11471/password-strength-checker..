```python
password = input("Enter your password: ")

length = len(password)

has_number = False
has_upper = False
has_special = False

for ch in password:
    if ch.isdigit():
        has_number = True
    elif ch.isupper():
        has_upper = True
    elif not ch.isalnum():
        has_special = True

if length >= 8 and has_number and has_upper and has_special:
    print("Password is Strong")
elif length >= 6 and (has_number or has_upper):
    print("Password is Medium")
else:
    print("Password is Weak")
```
